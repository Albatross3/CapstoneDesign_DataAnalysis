# Normal, Botnet data만 남기기


```python
index=dataset[dataset['Label'].str.contains('Background')].index # 'Background' 포함하는 index 찾기
dataset= dataset.drop(index)
dataset.reset_index(drop=True, inplace=True) # index 차례로 만들기
dataset
```

# Filling Null values

```python
# Sport, Dport 의 Na 값들은 따로 범주화 하므로 안채워도 됨
dataset.isnull().sum()
```

```python
# fill nan port forward (앞에 있는 data로 채우기)
dataset['sTos'] = dataset.sTos.fillna(method='pad')
dataset['dTos'] = dataset.dTos.fillna(method='pad')
```

## Feature engineering
필요없는 feature 제거 및 OneHotEncoding
1. StartTime, SrcAddr, DstAddr 제거
2. Protocol, sTos, dTos, dir -> OneHotEncoding 으로 변환
3. Sport, Dport ? -> github:16진법 변환, **pdf:Category 범주 축소**
4. State ? -> github: 없앰, **pdf:Category 범주 축소**

```python
# StartTime, SrcAddr, DstAddr 제거
dataset.drop(['StartTime','SrcAddr','DstAddr'],axis=1,inplace=True) # inplace -> dataframe 이름은 유지
```

```python
# port feature -> 널 값, 16진수, 0~1023, 1024~49151, 49152~65535
# Sport feature -> 범주화 ('Sempty','Snosp','Swell','Sregi','Spriv')
dataset['Sport'] = dataset.Sport.fillna(value='Sempty') # null 값 'Sempty' 로 채우기
dataset.loc[dataset['Sport'].str.startswith("0x"),'Sport']="Snosp"# 16진수 'Snosp' 로 채우기
```

```python
# Sport 숫자 범주화
for index in dataset.index:
  value=dataset.loc[index,'Sport']
  if value=="Sempty" or value=="Snosp": continue
  value=int(value)
  if value < 1024:
    dataset.loc[index,'Sport']="Swell"
  elif value < 49252:
    dataset.loc[index,'Sport']="Sregi"
  else:
    dataset.loc[index,'Sport']="Spriv"
```

```python
# Dport feature -> 범주화 ('Dempty','Dnosp','Dwell','Dregi','Dpriv')
dataset['Dport'] = dataset.Dport.fillna(value='Dempty') # null 값 'Dempty' 로 채우기
dataset.loc[dataset['Dport'].str.startswith("0x"),'Dport']="Dnosp"# 16진수 'Dnosp' 로 채우기
```

```python
# Dport 숫자 범주화
for index in dataset.index:
  value=dataset.loc[index,'Dport']
  if value=="Dempty" or value=="Dnosp": continue
  value=int(value)
  if value < 1024:
    dataset.loc[index,'Dport']="Dwell"
  elif value < 49252:
    dataset.loc[index,'Dport']="Dregi"
  else:
    dataset.loc[index,'Dport']="Dpriv"
```

```python
# State 범주 줄이기 -> RST
# RnR, Rn, n, s 크게 4가지
for index in dataset.index: 
  value=dataset.loc[index,'State']
  if not '_' in value:
    dataset.loc[index,'State']='s'
  else:
    temp=value.split('_') # list
    if 'R' in temp[0] and 'R' in temp[1]:
      dataset.loc[index,'State']='RnR'
    elif 'R' in temp[0]:
      dataset.loc[index,'State']='Rn'
    else:
      dataset.loc[index,'State']='n'
```

```python
# Proto, Sport, Dir, Dport, State, sTos, dTos 모두 one-hot-encoder 로 변환
from sklearn.preprocessing import OneHotEncoder
# Proto
ohe_proto = OneHotEncoder(sparse=False)
ohe_proto.fit(dataset[['Proto']])
df_proto = pd.DataFrame(ohe_proto.transform(dataset[['Proto']]), columns=['proto0', 'proto1', 'proto2','proto3'])

# Sport
ohe_sport = OneHotEncoder(sparse=False)
ohe_sport.fit(dataset[['Sport']])
df_sport = pd.DataFrame(ohe_sport.transform(dataset[['Sport']]), columns=['sport0', 'sport1', 'sport2','sport3','sport4'])

# Dir
ohe_dir = OneHotEncoder(sparse=False)
ohe_dir.fit(dataset[['Dir']])
df_dir = pd.DataFrame(ohe_dir.transform(dataset[['Dir']]), columns=['dir0','dir1','dir2','dir3','dir4'])

# Dport
ohe_dport = OneHotEncoder(sparse=False)
ohe_dport.fit(dataset[['Dport']])
df_dport = pd.DataFrame(ohe_dport.transform(dataset[['Dport']]), columns=['dport0', 'dport1', 'dport2','dport3','dport4'])

# State
ohe_state = OneHotEncoder(sparse=False)
ohe_state.fit(dataset[['State']])
df_state = pd.DataFrame(ohe_state.transform(dataset[['State']]), columns=['state0', 'state1', 'state2','state3'])

# sTos
ohe_stos = OneHotEncoder(sparse=False)
ohe_stos.fit(dataset[['sTos']])
df_stos = pd.DataFrame(ohe_stos.transform(dataset[['sTos']]), columns=['stos0'])

# dTos
ohe_dtos = OneHotEncoder(sparse=False)
ohe_dtos.fit(dataset[['dTos']])
df_dtos = pd.DataFrame(ohe_dtos.transform(dataset[['dTos']]), columns=['dtos0'])
```

```python
# ONE-HOT-ENCODING 한 부분 이어 붙이기
dataset = pd.concat([dataset[['Dur']], df_proto, df_sport, df_dir, df_dport, df_state, df_stos, df_dtos,  dataset[['TotPkts','TotBytes','SrcBytes','Label']]], axis=1)
```
