# Skoltech Anomaly Benchmark (SKAB) teaser
kaggle에 있는 센서 데이터

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
raw_data=pd.read_csv('dataset/SkAB teaser.csv',sep=';',index_col='datetime',parse_dates=True).drop('index',axis=1)
```

```python
raw_data.head(10)
```

```python
# pandas 피벗 테이블 함수 (특정 행에 따라서 data를 재배치할 때 사용)
raw_data = raw_data.pivot_table(values='value', index=raw_data.index,columns='id')
```

```python
raw_data.head()
```

```python
raw_data.info()
```

```python
raw_data.plot(figsize=(12,6), marker='o',markersize=3);
```

```python
# 중간에 잘린 부분 없애기 위해서 data 자름
raw_data =raw_data['2019-07-08 17:52:29':]
```

```python
raw_data.plot(figsize=(12,6), marker='o',markersize=3);
```

```python
# data의 개수
print(f'data의 크기: {raw_data.shape}')
```

```python
raw_data.describe()
```

```python
# 모든 신호의 그래프 그리기
for name in raw_data.columns:
    raw_data[name].plot(figsize=(12,3), marker='o',markersize=2)
    plt.xlabel('time')
    plt.ylabel('parameter value')
    plt.title(f'signal graph: {name}')
    plt.show()
```

# 모델 교육
대부분의 기계학습 알고리즘에서 data가 동일한 규모로 있어야 한다

```python
data=raw_data.copy()
```

```python
# 훈련,검증,테스트 데이터로 나눔
# 18:25 전, 18:25-18:35, 18:35 후 로 각각 훈련,검증,테스트
StSc= StandardScaler() 
StSc.fit(data[:'2019-07-08 18:25'])

train_sc= StSc.transform(data[:'2019-07-08 18:25']) #train_sc 는 numpy로 바뀜
```

```python
# time의 길이
len(train_sc)
```

```python
#np.corrcoef()
corrcoef=[]
for i in range(len(train_sc)-1):
    corrcoef.append(np.corrcoef(train_sc[i],train_sc[i+1])[0,1])
```

```python
plt.figure(figsize=(12,3))
plt.plot(corrcoef[0:100])
plt.xlabel('time')
plt.ylabel('correlation coefficient')
plt.show()

## coefficient 방법에 문제가 있는듯함..
```
